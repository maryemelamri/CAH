```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# Coordonnées des points
points = np.array([[1, 3], [2, 2], [2, 3], [2, 4], [4, 2], [5, 2], [6, 2], [7, 3]])

# Calcul de la matrice de liaison (linkage matrix) avec la méthode de Ward
linkage_matrix = linkage(points, 'ward')

# Affichage du dendrogramme
dendrogram(linkage_matrix, labels=range(1, len(points) + 1), leaf_rotation=90)
plt.title('Dendrogramme de la Classification Ascendante Hiérarchique')
plt.xlabel('Points')
plt.ylabel('Distance Euclidienne')

# Seuil de coupure pour former des clusters
threshold = 3.5

# Récupérer les clusters
clusters = fcluster(linkage_matrix, threshold, criterion='distance')

# Afficher les clusters obtenus
print("Clusters obtenus avec le seuil de coupure:", clusters)

plt.show()
```

```python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt

# Définir la matrice
X = np.array([1, 2, 9, 12, 20]).reshape(-1, 1)

# Calculer les liens avec la méthode de Ward
Z = linkage(X, 'ward')

# Afficher le dendrogramme
dendrogram(Z)

plt.title('Dendrogramme CAH - Méthode de Ward')
plt.ylabel('Distance euclidienne')

# Couper l'arbre en utilisant la méthode du saut minimal
threshold = 4  # Ajustez ce seuil en fonction de votre cas
clusters = fcluster(Z, t=threshold, criterion='distance')

plt.axhline(y=threshold, color='black', linestyle='--', label=f'Seuil de coupure = {threshold}')
plt.legend()

plt.show()

# Afficher les clusters obtenus
print("Clusters obtenus avec le seuil de coupure:", clusters)
```
